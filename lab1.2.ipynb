# Подготовка данных и импорт библиотек
Импорт необходимых библиотек: numpy, pandas, matplotlib, sklearn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Загрузка и анализ данных
Загрузка датасетов X и Y, вывод статистических характеристик и визуализация данных

```python
# Загрузка и анализ данных

# Загрузка данных
x = pd.read_csv("https://github.com/koroteevmv/ML_course/raw/main/ML1.2%20multivariate/data/0_x.csv", header=None)
y = pd.read_csv("https://github.com/koroteevmv/ML_course/raw/main/ML1.2%20multivariate/data/0_y.csv", header=None)

# Вывод первых строк датасета
print("Первые строки датасета X:")
print(x.head())
print("\nПервые строки датасета Y:")
print(y.head())

# Количественное описание датасета
print("\nРазмер датасета X:", x.shape)
print("Размер датасета Y:", y.shape)

# Статистика по каждому признаку
print("\nСтатистика по признакам датасета X:")
print(x.describe())

# Визуализация данных
plt.figure(figsize=(12, 8))
for i in range(x.shape[1]):
    plt.subplot(2, 3, i + 1)
    plt.scatter(x[i], y, alpha=0.5)
    plt.xlabel(f'Признак {i+1}')
    plt.ylabel('Целевая переменная')
    plt.title(f'Признак {i+1} vs Целевая переменная')
plt.tight_layout()
plt.show()
```

# Построение парных регрессий
Создание и обучение моделей линейной регрессии для каждого признака отдельно

```python


for i in range(x.shape[1]):
    # Создание модели линейной регрессии
    model = LinearRegression()
    
    # Обучение модели на одном признаке
    model.fit(x[[i]], y)
    
    
    y_pred = model.predict(x[[i]])
    
   
    plt.figure(figsize=(8, 6))
    plt.scatter(x[i], y, alpha=0.5, label='Исходные данные')
    plt.plot(x[i], y_pred, color='red', label='Линия регрессии')
    plt.xlabel(f'Признак {i+1}')
    plt.ylabel('Целевая переменная')
    plt.title(f'Парная регрессия: Признак {i+1} vs Целевая переменная')
    plt.legend()
    plt.show()
    
   
    print(f'Признак {i+1}:')
    print('Коэффициенты регрессии:', model.coef_)
    print('Свободный член:', model.intercept_)
    print('Коэффициент детерминации (R^2):', model.score(x[[i]], y))
    print('\n')
```

# Построение множественной регрессии
Создание и обучение модели множественной линейной регрессии со всеми признаками

```python


# Создание модели множественной линейной регрессии
multiple_model = LinearRegression()


multiple_model.fit(x, y)

y_pred_multiple = multiple_model.predict(x)

plt.figure(figsize=(8, 6))
plt.scatter(y, y_pred_multiple, alpha=0.5, label='Предсказанные значения')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', label='Идеальная линия')
plt.xlabel('Истинные значения')
plt.ylabel('Предсказанные значения')
plt.title('Множественная регрессия: Истинные vs Предсказанные значения')
plt.legend()
plt.show()

print('Коэффициенты регрессии:', multiple_model.coef_)
print('Свободный член:', multiple_model.intercept_)
print('Коэффициент детерминации (R^2):', multiple_model.score(x, y))
```

# Построение полиномиальных регрессий
Создание полиномиальных признаков и обучение моделей разных степеней

```python
from sklearn.preprocessing import PolynomialFeatures



degrees = [2, 3, 10]

for degree in degrees:
    print(f'\nПолиномиальная регрессия степени {degree}:\n')
    
    poly = PolynomialFeatures(degree)
    x_poly = poly.fit_transform(x)
    poly_model = LinearRegression()
    poly_model.fit(x_poly, y)
    y_pred_poly = poly_model.predict(x_poly)
    plt.figure(figsize=(8, 6))
    plt.scatter(y, y_pred_poly, alpha=0.5, label='Предсказанные значения')
    plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', label='Идеальная линия')
    plt.xlabel('Истинные значения')
    plt.ylabel('Предсказанные значения')
    plt.title(f'Полиномиальная регрессия степени {degree}: Истинные vs Предсказанные значения')
    plt.legend()
    plt.show()

    print('Коэффициенты регрессии:', poly_model.coef_)
    print('Свободный член:', poly_model.intercept_)
    print('Коэффициент детерминации (R^2):', poly_model.score(x_poly, y))
```

# Сравнение метрик качества моделей
Расчет и сравнение различных метрик качества для всех построенных моделей

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


def print_metrics(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f'{model_name}:')
    print(f'Среднеквадратичная ошибка (MSE): {mse}')
    print(f'Средняя абсолютная ошибка (MAE): {mae}')
    print(f'Коэффициент детерминации (R^2): {r2}')
    print('\n')

for i in range(x.shape[1]):
    model = LinearRegression()
    model.fit(x[[i]], y)
    y_pred = model.predict(x[[i]])
    print_metrics(y, y_pred, f'Парная регрессия: Признак {i+1}')

y_pred_multiple = multiple_model.predict(x)
print_metrics(y, y_pred_multiple, 'Множественная регрессия')

for degree in degrees:
    poly = PolynomialFeatures(degree)
    x_poly = poly.fit_transform(x)
    poly_model = LinearRegression()
    poly_model.fit(x_poly, y)
    y_pred_poly = poly_model.predict(x_poly)
    print_metrics(y, y_pred_poly, f'Полиномиальная регрессия степени {degree}')
```

# Анализ времени обучения
Замер и визуализация времени обучения разных моделей

```python
import time

training_times = []

# Время обучения парных регрессий
for i in range(x.shape[1]):
    start_time = time.time()
    model = LinearRegression()
    model.fit(x[[i]], y)
    end_time = time.time()
    training_times.append(['Признак ' + str(i+1), end_time - start_time])

# Время обучения множественной регрессии
start_time = time.time()
multiple_model = LinearRegression()
multiple_model.fit(x, y)
end_time = time.time()
training_times.append(['Множественная регрессия', end_time - start_time])

# Время обучения полиномиальных регрессий
for degree in degrees:
    start_time = time.time()
    poly = PolynomialFeatures(degree)
    x_poly = poly.fit_transform(x)
    poly_model = LinearRegression()
    poly_model.fit(x_poly, y)
    end_time = time.time()
    training_times.append([f'Полиномиальная регрессия степени {degree}', end_time - start_time])

training_times_df = pd.DataFrame(training_times, columns=['Модель', 'Время обучения (секунды)'])
print(training_times_df)

plt.figure(figsize=(10, 6))
plt.barh(training_times_df['Модель'], training_times_df['Время обучения (секунды)'], color='skyblue')
plt.xlabel('Время обучения (секунды)')
plt.ylabel('Модель')
plt.title('Время обучения различных моделей регрессии')
plt.show()
```
